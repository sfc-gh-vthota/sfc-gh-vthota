## Part 1: Storage Lifecycle Policies (Data Archival)

Storage Lifecycle Policies automatically move data between storage tiers based on age:
- **ACTIVE** tier - Full performance, highest cost
- **COOL** tier - Reduced cost (~40% savings), slightly higher access latency
- **COLD** tier - Lowest cost, highest access latency (for rarely accessed data)

### Use Case: Banking Regulatory Retention
Banks must retain transaction data for 7+ years. Active analytics only needs recent data. Archive older data to save costs while maintaining compliance.

````python
-- =============================================================================
-- STORAGE LIFECYCLE POLICY: Archive transaction data after 90 days
-- =============================================================================

-- Create a lifecycle policy
CREATE OR REPLACE STORAGE LIFECYCLE POLICY HOL_TXN_ARCHIVE_POLICY
    COMMENT = 'Archive transactions older than 90 days to COOL tier'
AS
    -- Move to COOL tier after 90 days
    ARCHIVE_FOR_DAYS = 90;
````

````python
-- Attach the lifecycle policy to the TRANSACTIONS table
ALTER TABLE TRANSACTIONS 
    SET STORAGE_LIFECYCLE_POLICY = HOL_TXN_ARCHIVE_POLICY
    STORAGE_LIFECYCLE_POLICY_DATE_COLUMN = 'TXN_DATE';

-- Verify policy attachment
SHOW STORAGE LIFECYCLE POLICIES;

-- Check which tables have lifecycle policies
SELECT TABLE_NAME, STORAGE_LIFECYCLE_POLICY
FROM INFORMATION_SCHEMA.TABLES
WHERE TABLE_SCHEMA = 'HOL_LAB'
AND STORAGE_LIFECYCLE_POLICY IS NOT NULL;
````

````python
-- Query archived data (transparent access - same SQL, just slightly higher latency)
-- This query reads from archived partitions seamlessly:
SELECT 
    DATE_TRUNC('MONTH', TO_TIMESTAMP(TXN_DATE)) AS TXN_MONTH,
    COUNT(*) AS TXN_COUNT,
    SUM(AMOUNT) AS TOTAL_AMOUNT
FROM TRANSACTIONS
WHERE TO_TIMESTAMP(TXN_DATE) < DATEADD('DAY', -90, CURRENT_TIMESTAMP())
GROUP BY TXN_MONTH
ORDER BY TXN_MONTH;

-- Note: Archived data is still queryable! Same SQL, same semantics.
-- The only difference is slightly higher latency for COOL/COLD tier access.
````

## Part 2: Iceberg Table Interoperability

Our managed Iceberg tables (ACCOUNTS, LOANS, CREDIT_CARDS) are stored in Apache Iceberg open format. This means:
- External engines (Spark, Trino, Flink) can read them directly
- No data copy or export needed
- Snowflake's Horizon Catalog exposes metadata via REST API

````python
-- =============================================================================
-- ICEBERG TABLE METADATA & INTEROPERABILITY
-- =============================================================================

-- View Iceberg table information (metadata, snapshots, manifests)
SELECT SYSTEM$GET_ICEBERG_TABLE_INFORMATION('ACCOUNTS') AS ICEBERG_INFO;
````

````python
-- View Iceberg table properties
SHOW ICEBERG TABLES IN SCHEMA HOL_LAB;

-- Check table metadata including snapshot info
SELECT 
    TABLE_NAME,
    TABLE_TYPE,
    IS_ICEBERG,
    CATALOG,
    BASE_LOCATION
FROM INFORMATION_SCHEMA.TABLES 
WHERE TABLE_SCHEMA = 'HOL_LAB' 
AND IS_ICEBERG = 'YES';
````

````python
-- =============================================================================
-- CATALOG INTEGRATION (Snowflake-managed catalog for external access)
-- =============================================================================

-- Create a catalog integration that external engines can use
-- This exposes the Iceberg REST Catalog endpoint for Spark/Trino
CREATE OR REPLACE CATALOG INTEGRATION HOL_ICEBERG_CATALOG
    CATALOG_SOURCE = SNOWFLAKE
    TABLE_FORMAT = ICEBERG
    ENABLED = TRUE
    COMMENT = 'Catalog integration for external engine access to managed Iceberg tables';

-- Show the catalog integration details (includes REST endpoint)
DESCRIBE CATALOG INTEGRATION HOL_ICEBERG_CATALOG;
````

````python
-- Demonstrate that Iceberg tables support standard Snowflake operations
-- (they work just like native tables from Snowflake's perspective)

-- Time travel on Iceberg tables
SELECT COUNT(*) FROM ACCOUNTS AT(OFFSET => -60);  -- 60 seconds ago

-- DML on Iceberg tables
UPDATE ACCOUNTS 
SET BALANCE = BALANCE * 1.001  -- Apply tiny interest
WHERE STATUS = 'ACTIVE'
AND ACCOUNT_TYPE = 'SAVINGS'
LIMIT 100;

-- View snapshot history (each DML creates a new snapshot)
SELECT SYSTEM$GET_ICEBERG_TABLE_INFORMATION('ACCOUNTS');
````
